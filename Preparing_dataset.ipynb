```python
import os
import shutil
import random
from osgeo import gdal
```

```python
path = "MyDrive/SGP7"
SEED = random.seed(123)
```

```python
categories = []
tif_files = []
for (dirpath, dirnames, filenames) in os.walk(path + '/2750'):
    categories.extend(dirnames)
    tif_files += filenames
print(categories)
print(tif_files)
```

```python
split_names = ['train', 'test', 'valid']
for sp_name in split_names:
    directory = path + '/2750' + sp_name + '/'
    if not os.path.exists(directory):
        os.makedirs(directory)
    # make category folders
    for category in categories:
        dir_cat = directory + category + '/'
        if not os.path.exists(dir_cat):
            os.makedirs(dir_cat)
```

```python
print(categories)
for category in categories:
  directory = path + '/2750/' + category + '/'
  directory_train = path + '/2750train/' + category + '/'
  directory_valid = path + '/2750valid/' + category + '/' 
  directory_test = path + '/2750test/' + category + '/'
  
  cat_files = os.listdir(directory)
  print(cat_files) 
  if '.DS_Store' in cat_files:
      cat_files.remove('.DS_Store')
  
  filenames = os.listdir(directory)
  filenames.sort()
  if '.DS_Store' in filenames:
    filenames.remove('.DS_Store')
  random.shuffle(filenames)
  print(filenames)
  split_1 = int(0.8 * len(filenames))
  split_2 = int(0.9 * len(filenames))
  train_filenames = filenames[:split_1]
  valid_filenames = filenames[split_1:split_2]
  test_filenames = filenames[split_2:]
  print('hiiiiii')      
  for file in filenames:
      if file in valid_filenames:
          shutil.move(directory + file, directory_valid + file)
          print('success')
      elif file in test_filenames:
          shutil.move(directory + file, directory_test + file)
```
